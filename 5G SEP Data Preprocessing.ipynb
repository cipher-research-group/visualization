## 5G SEP Data Preprocessing

### Import Libraries

```python
import pandas as pd
from collections import OrderedDict 
```

### Load the Dataset

```python
patent_data = pd.read_excel('5G_SEP_Data_All.xlsx')
```

```python
patent_data
```

```python
# Select only the important fields
patent_data = patent_data.loc[:,['Earliest application date', 'Current assignees']]
```

```python
patent_data
```

```python
# Grab Only the 'Earliest application date' from the date
for idx, i in enumerate(patent_data['Earliest application date']):
    #print(i)
    patent_data.loc[idx,'Earliest application date'] = int(i.split('-')[0])
    # Clean Company Names
    if "\r\n" in patent_data.loc[idx,'Current assignees']:
        patent_data.loc[idx,'Current assignees'] = patent_data.loc[idx,'Current assignees'].replace("\r\n", " and ")
        
    
```

```python
patent_data
```

### Dictionary - name: year: count 

```python
patent_dict = {}

for i in range(0,len(patent_data)):
    if patent_data.loc[i,'Current assignees'] not in patent_dict:
        patent_dict[patent_data.loc[i,'Current assignees']] = {}
        if patent_data.loc[i,'Earliest application date'] not in patent_dict[patent_data.loc[i,'Current assignees']]:
            patent_dict[patent_data.loc[i,'Current assignees']][patent_data.loc[i,'Earliest application date']] = 1
        elif patent_data.loc[i,'Earliest application date'] in patent_dict[patent_data.loc[i,'Current assignees']]:
            patent_dict[patent_data.loc[i,'Current assignees']][patent_data.loc[i,'Earliest application date']] += 1
    elif patent_data.loc[i,'Current assignees'] in patent_dict:
        if patent_data.loc[i,'Earliest application date'] not in patent_dict[patent_data.loc[i,'Current assignees']]:
            patent_dict[patent_data.loc[i,'Current assignees']][patent_data.loc[i,'Earliest application date']] = 1
        elif patent_data.loc[i,'Earliest application date'] in patent_dict[patent_data.loc[i,'Current assignees']]:
            patent_dict[patent_data.loc[i,'Current assignees']][patent_data.loc[i,'Earliest application date']] += 1
    
    
```

```python
for i in patent_dict.keys():
    patent_dict[i] = dict(sorted(patent_dict[i].items(), reverse = True))
```

### Convert Dictionary to Pandas Dataframe: if you want non cumulative data

```python
patent_dict_data = pd.DataFrame(columns=['name','year','value','lastValue','group'])
count=0
last_value=0
for idx, name in enumerate(patent_dict):
    counter_new_court = 0
    for year in patent_dict[name]:
        count+=1
        counter_new_court+=1
        patent_dict_data.loc[count,'name'] = name
        patent_dict_data.loc[count,'group'] = name
        patent_dict_data.loc[count,'year'] = year
        patent_dict_data.loc[count,'value'] = patent_dict[name][year]
        
        if counter_new_court==0:
            patent_dict_data.loc[count,'lastValue'] = patent_dict[name][year]
        elif counter_new_court > 0:
            patent_dict_data.loc[count,'lastValue'] = last_value
            
        last_value = patent_dict_data.loc[count,'value']
        
    
```
